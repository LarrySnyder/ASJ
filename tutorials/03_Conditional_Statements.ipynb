# Conditional Statements

Lots of times in computer programs, we need to do one thing or another based on whether some condition is true or false.

For example:
* If a variable `i` is greater than 10, divide it by 2.
* Otherwise, add 1 to it.

Or:
* If the user checked a certain checkbox, display the current time.
* Otherwise, don't.

Or:
* If `gender` equals `'Female'`, print `'she'`.
* Otherwise, if `gender` equals `'Male'`, print `'he'`.
* Otherwise, print `'they'`.

We check conditions using **conditional statements**—in particular, `if` statements.

## `if` Statements



Let's check whether a variable `i` is greater than 10; if so, we'll divide it by 2, and if not, we'll leave it as is:

```
if z > 10:
  z /= 2
else:
  z += 1
print(z)
```

Python checks the condition that comes after the `if`. If the condition is true, Python executes the (indented!) code that comes next. Otherwise, it executes the (indented!) code that comes after the `else` statement.

If you execute the code above, you'll get an error, because you haven't yet created a variable called `z`. (Try it, if you want.) 

So add the following line *before* the code above:

```
z = 7
```

Now try it.

````python
z = 7
if z > 10:
  z /= 2
else:
  z += 1
print(z)
````

Try changing the `z = 7` line to `z = 15` and re-running the code. Are the results what you expect?

---

You don't *need* an `else`. For example:

```
x = 5
if x == 9:
  print("Now x = 9")
```

*Important*: To check whether a variable *equals* something, use *two* equal signs: `if x == 9`. This is how Python distinguishes between variable assignment and checking a condition.
* To *assign* a value to a variable, use one equal sign: `x = 5`
* To check *whether* a variable equals a value, use two equal signs: `x == 9`

## `elif` Statements



The keyword `elif` means "else if". It is used to check second or third or subsequent conditions after the initial `if` statement. It can be followed by an `else` statement (but doesn't have to be).

```
color_code = "g"
if color_code == "r":
  color = "red"
elif color_code == "g":
  color = "green"
elif color_code == "b":
  color = "blue"
else:
  print("Sorry, I don't recognize that color code.")
  color = None
print(f"The color is {color}.")
```

Note that `else` statement is followed by multiple indented commands. That's fine. You can execute multiple commands after `if`, `elif`, or `else`, just like you can inside of a loop.

The word `None` on the last line is a special keyword in Python. Here, it's like we are setting `color` to nothing. It's different from not setting `color` to anything! This is a bit of an advanced concept in Python, so let's not discuss it more it right now.

Try the code above. Then change the `"g"` in the first line to `"r"` or `"b"` or `"foo"` and make sure the code performs as expected.

````python
color_code = "g"
if color_code == "r":
  color = "red"
elif color_code == "g":
  color = "green"
elif color_code == "b":
  color = "blue"
else:
  print("Sorry, I don't recognize that color code.")
  color = None
print(f"The color is {color}.")
````

## Boolean Variables

*Boolean* variables can only take one of two values: `True` or `False`. (`True` and `False` are special Python keywords, and capitalization matters.)

We don't need comparison operators to check whether Boolean variables are true or false; for example:

```
user_checked_box = True
if user_checked_box:
  print("The user checked the checkbox.")
```

``if user_checked_box`` is equivalent to ``if user_checked_box == True``.

## `and`, `or`, and `not`



Often we need to combine several conditions into one. For example:
* Check whether `x` ≥ 10 and < 20
* Check whether `y` < 5 or `user_clicked_cancel` is `False`
* Check whether `b` > 10 but does not equal 14

The Python keywords `and`, `or`, and `not` are just the things we need. 

* `x >= 10 and x < 20`
* `y < 5 or not user_clicked_cancel`
* `z > 10 and not b == 14`

Note that `not` comes *before* the condition: We write `not b == 14`, instead of `b not == 14`. Of course, we could have used `b != 14`, but we chose `not` to. 



Sometimes, we need parentheses to group more complicated conditions. For example, suppose a university wants to admit students based on a formula that involves their SAT score, the number of extracurricular activities they participate in, and their GPA. Consider the following code:

```
if sat_score > 1300 and number_extracurriculars >= 3 or gpa > 3.5:
  admitted = True
```

This condition is ambiguous: Do we want students who have good SAT scores and *either* lots of extracurriculars or high GPA? Or do we want students who have *both* good SAT scores *and* lots of extracurriculars, or a high GPA? 

If a student has a 1200 SAT, 2 extracurriculars, and a 3.8 GPA, should they be admitted? 

Adding parentheses clears up the ambiguity:

```
if (sat_score > 1300 and number_extracurriculars >= 3) or gpa > 3.5:
  admitted = True
```

Now we know: We want students *both* good SAT scores *and* lots of extracurriculars, or a high GPA. Our 1200-SAT, 2-extracurricular, 3.8-GPA student should not be admitted.


## Nested `if` Statements

You can **nest** `if` statements inside other `if` statements, just like you can nest loops inside other loops.

Suppose morning meetings are 30 minutes long if they are remote and 45 minutes long if they are in person; and afternoon meetings are 15 minutes longer in either case. The following code determines the length of a meeting based on `start_time` (which tells the hour of the day at which the meeting starts) and `is_remote` (a Boolean variable). 

```
start_time = 18
is_remote = True
if start_time <= 12:
  if is_remote:
    meeting_length = 30
  else:
    meeting_length = 45
else:
  if is_remote:
    meeting_length = 45
  else:
    meeting_length = 60
print(f"The meeting will be {meeting_length} minutes long")
```

As always, the indents are critical so that Python knows how the nesting works.

Try it. Then try changing the values of the variables in the first two lines to make sure the results come out the way we expect them to.

````python
start_time = 18
is_remote = True
if start_time <= 12:
  if is_remote:
    meeting_length = 30
  else:
    meeting_length = 45
else:
  if is_remote:
    meeting_length = 45
  else:
    meeting_length = 60
print(f"The meeting will be {meeting_length} minutes long")
````

## `if` Statements Inside Loops

We often want to loop through a bunch of values and check a condition for each one. In that case, the `if` statement shows up inside a `for` or `while` loop.

The following code loops through the first 25 integers. For each one, it checks whether it's divisible by 3 or 7. If it is, it prints a message.

```
for n in range(25):
  if n % 3 == 0 or n % 7 == 0:
    print(f"{n} is divisible by either 3 or 7")
```

In Python, `%` means "remainder after dividing." So, if `n % 3` equals 0, that means `n` is divisible by 3.

Try it.

````python
for n in range(25):
  if n % 3 == 0 or n % 7 == 0:
    print(f"{n} is divisible by either 3 or 7")
````
