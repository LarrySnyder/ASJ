# Looping

In most computer programs, there are things we want to do multiple times. For example, suppose we want to print the squares of the integers 0 through 4. One way to do this is as follows:

```
print(0 ** 2)
print(1 ** 2)
print(2 ** 2)
print(3 ** 2)
print(4 ** 2)
```

**DON'T DO THIS!** This would be a silly way to print the first 5 squares. Programmers are lazy, and programming languages have evolved to coddle their laziness. If you find yourself doing (almost) the same thing multiple times in code, there is probably a better way (meaning, a way that requires less typing) to code it.

Instead, we should use a **loop** to print these squares. 

## `for` Loops

First, we will use a type of loop called a `for` loop.

Type the following into the cell below and execute it:

```
for i in range(5):
  print(i ** 2)
```

Note the indent before the `print` command. This is important! Indents (or lack thereof) are how Python knows what's inside the loop and what's outside. 

When you press <kbd>`Enter`</kbd> after the colon on the first line, the notebook will probably already indent the next line for you. Alternatively, you can get an indent using one <kbd>`Tab`</kbd> or two <kbd>`Space`</kbd>s. (But you can't mix and match: If you use tabs in one line, you can't use spaces in the next. Best to be consistent throughout your whole file.) The choice of spaces vs. tabs is up to the programmer and is the source of [much debate](https://youtu.be/SsoOG6ZeyUI).

````python
for i in range(5):
  print(i ** 2)
````

You just wrote a `for` loop!
 The structure works like this:
* The first line tells Python that you want to do a thing 5 times.
* The second line tells Python the thing you want to do. 

In particular:
* The `for i` part says "Create a variable called `i`." 
* You can think of `range(5)` as a list of integers from 0 to 4. (You tell it the first number you *don't* want, rather than the last number you *do* want.)
* The colon at the end of the first line tells Python, "here comes the stuff I want you to repeat."
* At each **iteration** of the loop, Python runs the part "inside" the loop and then increases `i` to the next number, then goes back to the start of the loop.
* The loop stops after 5 iterations.




If you want to print the first 5 squares, your `for` loop is a much better way to do it than using 5 separate `print` commands.

For one thing, it uses 2 lines of code instead of 5. Maybe you don't mind writing 5 lines, but what if we needed to print the first 50 or 500 or 5000 squares? You'd need 50 or 500 or 5000 separate `print` commands—not fun. But the `for` loop is still just 2 lines long; you just need to change the number inside the `range`. (Try it if you like.)

Another reason why the `for` loop is better than the multiple-`print` approach is that if you change your mind about *what* you want to print, you only need to change it once. For example, if we wanted to print cubes instead of squares, you just need to change the `2` to a `3` inside the single `print` command, as opposed to changing `2` to `3` in 5 separate `print` commands.

You can even put loops inside of other loops. These are called **nested** loops. For example:

```
for i in range(4):
  for j in range(4):
    print(f"{i} + {j} = {i + j}")
```

This loop prints the sum of each pair of integers that are between 0 and 4. Note that the `print` command is indented *twice*, once to show that it is inside of the `for j` loop, and once because the `for j` loop is itself indented.

The stuff inside the `print` command is a special Python syntax called an **f-string**. If you put the *name* of a variable inside curly braces (like `{i}`), Python will replace that expression with the *value* of the variable before it prints. 

So, if `i` = 2 and `j` = 1, the command `print(f"{i} + {j} = {i + j}")` will print 

```
2 + 1 = 3
```

Try executing the nested for loop.


````python
for i in range(4):
  for j in range(4):
    print(f"{i} + {j} = {i + j}")
````

## `while` Loops

Next we'll print the first 5 squares using a different kind of loop, called a `while` loop.

Type the following into the cell below and execute it:

```
i = 0
while i < 5:
  print(i ** 2)
  i += 1
```

As with a `for` loop, the commands inside a `while` loop must be indented. 

````python
i = 0
while i < 5:
  print(i ** 2)
  i += 1
````

Here's how your `while` loop works. First, outside the loop, you created a variable called `i` and set it to 0. 

Next, the line `while i < 5:` says, "keep doing the stuff that comes next as long as `i` is less than 5; if `i` is greater than or equal to 5, stop."

The first line inside the loop prints a square. The second line tells Python to increment `i` by 1. Another way to write the same command is

```
i = i + 1
```

But Python, and other languages, have a shortcut, because programmers increment variables by 1 a lot. The shortcut is flexible; you can increment `i` by 47 using:

```
i += 47
```

And you can do subtraction, multiplication, or division similarly:

```
i -= 2
i *= 0.5
i /= 10
```

Anyway, your `while` loop therefore says:
* Start with `i = 0`
* Print `i` squared, and increment `i` by 1
* Keep going until `i` is greater than or equal to 5

---


There's nothing inherently better about a `for` loop vs. a `while` loop. The `for` loop for printing squares uses 2 lines while the `while` loop uses 4, but that's not too bothersome. There are situations when one type of loop is more convenient for one reason or another, so that's the type you'd choose.
