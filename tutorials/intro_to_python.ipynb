One really useful thing about variables is that you can change them. (That's why they're called *variables*!) Try this:

```

print(a)
```

(*Note*: In Python, an equal sign `=` means "*assign* the value on the right to the variable on the left." It does *not* mean "the variable on the left *is equal to* the value on the right.")

````python
a = 37
print(a)
````

Variables don't have to have single-character names. In fact, it's often better to use longer, more descriptive names:

```
radius = 5
pi = 3.14
area = pi * radius ** 2
print(area)
```

(`**` means "to the power of." Python's "order of operations" means that Python will first square `radius`, then multiply the result by `pi`, instead of first multiplying `pi` by `radius`, then squaring the result.)

Try it. 

````python
radius = 5
pi = 3.14
area = pi * radius ** 2
print(area)
````

Variable names can contain underscores (`_`) and digits but not dashes or other special characters. The following are all fine:

* `name_of_file`
* `number_of_rows`
* `my_variable_2`

You don't *need* the underscores. Python will *allow* you to create variables like `numberofrows`, but best-practices say that it's better to use underscores, for readability.
