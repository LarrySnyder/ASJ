# Conditional Statements

Lots of times in computer programs, we need to do one thing or another based on whether some condition is true or false.

For example:
* If a variable `i` is greater than 10, divide it by 2.
* Otherwise, add 1 to it.

Or:
* If the user checked a certain checkbox, display the current time.
* Otherwise, don't.

Or:
* If `gender` equals `'Female'`, print `'she'`.
* Otherwise, if `gender` equals `'Male'`, print `'he'`.
* Otherwise, print `'they'`.

We check conditions using **conditional statements**—in particular, `if` statements.

## `if` Statements

Let's check whether a variable `i` is greater than 10; if so, we'll divide it by 2, and if not, we'll leave it as is:

```
if z > 10:
  z /= 2
else:
  z += 1
print(z)
```

Python checks the condition that comes after the `if`. If the condition is true, Python executes the (indented!) code that comes next. Otherwise, it executes the (indented!) code that comes after the `else` statement.

If you execute the code above, you'll get an error, because you haven't yet created a variable called `z`. (Try it, if you want.) 

So add the following line *before* the code above:

```
z = 7
```

Now try it.

````python
z = 7
if z > 10:
  z /= 2
else:
  z += 1
print(z)
````

Try changing the `z = 7` line to `z = 15` and re-running the code. Are the results what you expect?

---

You don't *need* an `else`. For example:

```
x = 5
if x == 9:
  print("Now x = 9")
```

*Important*: To check whether a variable *equals* something, use *two* equal signs: `if x == 9`. This is how Python distinguishes between variable assignment and checking a condition.
* To *assign* a value to a variable, use one equal sign: `x = 5`
* To check *whether* a variable equals a value, use two equal signs: `x == 9`

## `elif` Statements

The keyword `elif` means "else if". It is used to check second or third or subsequent conditions after the initial `if` statement. It can be followed by an `else` statement (but doesn't have to be).

```
color_code = "g"
if color_code == "r":
  color = "red"
elif color_code == "g":
  color = "green"
elif color_code == "b":
  color = "blue"
else:
  print("Sorry, I don't recognize that color code.")
  color = None
print(f"The color is {color}.")
```

Note that `else` statement is followed by multiple indented commands. That's fine. You can execute multiple commands after `if`, `elif`, or `else`, just like you can inside of a loop.

The word `None` on the last line is a special keyword in Python. Here, it's like we are setting `color` to nothing. It's different from not setting `color` to anything! This is a bit of an advanced concept in Python, so let's not discuss it more it right now.

Try the code above. Then change the `"g"` in the first line to `"r"` or `"b"` or `"foo"` and make sure the code performs as expected.

````python
color_code = "g"
if color_code == "r":
  color = "red"
elif color_code == "g":
  color = "green"
elif color_code == "b":
  color = "blue"
else:
  print("Sorry, I don't recognize that color code.")
  color = None
print(f"The color is {color}.")
````
